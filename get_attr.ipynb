```python
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
import math
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

ReadTxt = '/content/drive/MyDrive/DL/CelebA/Anno/list_attr_celeba.txt'
r = open(ReadTxt,'r')
total = int(r.readline().split('\n')[0]) #total number
readLine = r.readline().split('\n')[0] #the title
titleLine = np.array(readLine.split())


readLine = r.readline().split('\n')[0] #first line
attDist = np.zeros(40)
count = 0
while readLine and count < 15000:
    temp = np.array(readLine.split()[1:], dtype = int)

    attDist = attDist + temp
    count=count+1
    readLine = r.readline().split('\n')[0]
r.close()
```

```python
sortIndex = np.argsort(np.abs(attDist))
titleLine = titleLine[sortIndex]
arrDist = np.abs(attDist)[sortIndex]
```

```python
#print(titleLine)
plt.figure(figsize = (10,4))
plt.xlabel('attribute', fontsize = 10)
plt.xticks(rotation = 90, fontsize = 10)
plt.bar(titleLine[0:40], arrDist[0:40])
print(sortIndex[0:5])
```

```python
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

ReadTxt = '/content/drive/MyDrive/DL/CelebA/Anno/list_attr_celeba.txt'
WriteTxt = '/content/drive/MyDrive/DL/att_list.txt'
r = open(ReadTxt,'r')
w = open(WriteTxt,'w')
total = int(r.readline().split('\n')[0]) #total number
readLine = r.readline().split('\n')[0] #the title
wLine = readLine.split()
wLine = wLine[2]+' '+wLine[31]+' '+wLine[21]+' '+wLine[36]+' '+wLine[19]+'\n'
w.write(wLine)
readLine = r.readline().split('\n')[0] #first line
count = 0
while readLine:
#    File name, 'Attractive' 'Smiling' 'Mouth_Slightly_Open' 'Wearing_Lipstick' 'High_Cheekbones'
    wLine = readLine.split()
    wLine = ' '+str(count+1)+' '+wLine[3]+' '+wLine[32]+' '+wLine[22]+' '+wLine[37]+' '+wLine[20]+'\n'         
    w.write(wLine)
    count=count+1
    readLine = r.readline().split('\n')[0]
r.close()
w.close()

if total == count:
  print('All lable import successfully')
```
